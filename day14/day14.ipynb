```python
data = open('input.txt').read()[:-1].split('\n')
# data = '''498,4 -> 498,6 -> 496,6
# 503,4 -> 502,4 -> 502,9 -> 494,9'''.split('\n')
# data = '''2,4 -> 2,6'''.split('\n')
print(data[0])
```

```python
class Resevoir:
    def __init__(self, setup, add_floor):
        WIDTH = 600
        HEIGHT = 200
        self.has_floor = add_floor
        self.board = [[None]*WIDTH for _ in range(HEIGHT)]
        self.bounds = [WIDTH, 0]
        floor = 0
        for struct in setup:
            corners = [tuple(map(int, corner.split(','))) for corner in struct.split(' -> ')]
            prev = corners[0]
            self.board[prev[1]][prev[0]] = '#'
            self.bounds[0] = min(self.bounds[0], prev[0])
            self.bounds[1] = max(self.bounds[1], prev[0])
            floor = max(floor, prev[1])
            
            for corner in corners[1:]:
                self.bounds[0] = min(self.bounds[0], corner[0])
                self.bounds[1] = max(self.bounds[1], corner[0])
                floor = max(floor, corner[1])
                
                for x in range(min(prev[0], corner[0]), max(prev[0], corner[0])+1):
                    for y in range(min(prev[1], corner[1]), max(prev[1], corner[1])+1):
                        self.board[y][x] = '#'
                        
                prev = tuple(corner)
        if add_floor:
            self.board = self.board[:floor+2]
            self.board.append(['#']*WIDTH)
    
    def drop_sand(self, x) -> bool: # return if sand **LANDS**
        if self.board[0][x] == 'O':
            return False
        
        for y in range(len(self.board)-1):
            if x == 0:
                for i in range(len(self.board)-1):
                    self.board[i].insert(0, None)
                self.board[-1].append('#')
                x += 1
            elif x == len(self.board[0])-1:
                for i in range(len(self.board)-1):
                    self.board[i].append(None)
                self.board[-1].append('#')
            
            if self.board[y+1][x] == None:
                continue
            elif self.board[y+1][x-1] == None:
                x -= 1
            elif self.board[y+1][x+1] == None:
                x += 1
            else:
                self.board[y][x] = 'O'
                return True
        
        return False
    
    def dump_sand(self, x) -> int:
        amt = 0
        while self.drop_sand(x):
            amt += 1
        return amt

    def __str__(self):
        # we love shitty oneliners
        return '\n'.join(' '.join(point if point else '.' for point in row) for row in self.board) if self.has_floor else \
               '\n'.join(' '.join(point if point else '.' for point in row[self.bounds[0]-1:self.bounds[1]+2]) for row in self.board)
```

```python
rsvr = Resevoir(setup=data, add_floor=False)
# print(rsvr)
print(rsvr.dump_sand(x=500))
print(rsvr)
```
